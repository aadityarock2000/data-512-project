## Relation between respiratory based mortality with Smoke estimates and pollution

```python
import json
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
#from pmdarima import auto_arima
from statsmodels.tsa.arima.model import ARIMA
```

## Smoke Estimate Model

```python
with open('../data/Grand_Junction_Wildfire_Features_geometry_dropped.json') as f:
    my_features = json.load(f)
```

```python
# Finding the average smoke created each year.
total_smoke = {}
count_per_year = {}

for fire in my_features:
    year = fire['attributes']['Fire_Year']
    area = fire['attributes']['GIS_Hectares']
    distance = fire['attributes']['distance_from_city']
    #smoke_created = 1.2* (1- distance/1250)*(150-area/150) 
    smoke_created = ((10**5)*area)/distance**3 + 50
    
    if year in total_smoke:
        total_smoke[year] += smoke_created
        count_per_year[year] += 1
    else:
        total_smoke[year] = smoke_created
        count_per_year[year] = 1

# Calculate the average smoke created for each year
average_smoke = {year: total_smoke[year] / count_per_year[year] for year in total_smoke}
```

```python
years = list(average_smoke.keys())
values = list(average_smoke.values())

plt.figure(figsize=(12, 6))
plt.plot(years, values, marker='o', linestyle='-')
plt.title('Time Series Plot')
plt.xlabel('Year')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler

# Assuming 'other_male', 'average_smoke_area_distance_estimate', and 'average_top_5_aqi' are your DataFrames

# Normalizing 'mx' from 'other_male'
scaler = MinMaxScaler()
mx_normalized = scaler.fit_transform(other_female['mx'].values.reshape(-1, 1))
years_mx = other_female['year_id'].values.reshape(-1, 1)  # Reshape to a 2D array for regression

# Normalizing 'Smoke Estimate' from 'average_smoke_area_distance_estimate'
smoke_normalized = scaler.fit_transform(average_smoke_area_distance_estimate['Smoke Estimate'].values.reshape(-1, 1))
years_smoke = average_smoke_area_distance_estimate['Year'].values.reshape(-1, 1)  # Reshape to a 2D array for regression

# Normalizing 'Max AQI' from 'average_top_5_aqi'
max_aqi_normalized = scaler.fit_transform(aqi_data['Max AQI'].values.reshape(-1, 1))
years_max_aqi = aqi_data['Year'].values.reshape(-1, 1)  # Reshape to a 2D array for regression

# Perform linear regression for 'mx' after normalization
regression_mx = LinearRegression()
regression_mx.fit(years_mx, mx_normalized)
predicted_mx_normalized = regression_mx.predict(years_mx)

# Perform linear regression for 'Smoke Estimate' after normalization
regression_smoke = LinearRegression()
regression_smoke.fit(years_smoke, smoke_normalized)
predicted_smoke_normalized = regression_smoke.predict(years_smoke)

# Perform linear regression for 'Max AQI' after normalization
regression_max_aqi = LinearRegression()
regression_max_aqi.fit(years_max_aqi, max_aqi_normalized)
predicted_max_aqi_normalized = regression_max_aqi.predict(years_max_aqi)

# Plotting 'mx', 'Smoke Estimate', and 'Max AQI' over the years with regression lines
plt.figure(figsize=(8, 6))

# Plot normalized 'mx' over the years
plt.scatter(other_female['year_id'], mx_normalized, label='mx (Normalized) - Original data', color='blue')
plt.plot(other_female['year_id'], predicted_mx_normalized, color='blue', linestyle='--', label='mx (Normalized) - Regression line')

# Plot normalized 'Smoke Estimate' over the years
plt.scatter(average_smoke_area_distance_estimate['Year'], smoke_normalized, label='Smoke Estimate (Normalized) - Original data', color='green')
plt.plot(average_smoke_area_distance_estimate['Year'], predicted_smoke_normalized, color='green', linestyle='--', label='Smoke Estimate (Normalized) - Regression line')

# Plot normalized 'Max AQI' over the years
plt.scatter(aqi_data['Year'], max_aqi_normalized, label='Max AQI (Normalized) - Original data', color='red')
plt.plot(aqi_data['Year'], predicted_max_aqi_normalized, color='red', linestyle='--', label='Max AQI (Normalized) - Regression line')

plt.xlabel('Year')
plt.ylabel('Normalized Values')
plt.title('Normalized mx, Smoke Estimate, and Max AQI Over the Years with Regression Lines')
plt.legend()
plt.grid(True)
plt.show()
```
